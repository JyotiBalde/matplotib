```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
height=[67,78,89,109,123]
weight=[32,45,46,56,76]

plt.plot(height,weight)
```

```python
plt.plot(height,weight)
plt.title("Realationship between height and weight")
plt.xlabel("height")
plt.ylabel("weight")
```

```python
calories_burnt=[10,20,30,40,60]
plt.plot(calories_burnt)
plt.plot(weight)
```

```python
plt.plot(calories_burnt)
plt.plot(weight)
plt.xlabel("calories_burnt")
plt.ylabel("weight")
plt.legend(labels=["calories_burnt","weight"],loc="best")
```

```python
plt.plot(calories_burnt,marker='o')
plt.plot(weight,marker='*')
plt.xlabel("calories_burnt")
plt.ylabel("weight")
plt.legend(labels=["calories_burnt","weight"],loc="lower right")
plt.xticks(ticks=[0,1,2,3,4],labels=["p1","p2","p3","p4","p5"]);
```

```python
#figure size in inches
plt.figure( figsize=[7,5])
plt.plot(calories_burnt)
plt.plot(weight)
plt.xlabel("calories_burnt")
plt.ylabel("weight")
plt.legend(labels=["calories_burnt","weight"],loc="best")
```

```python
plt.plot(calories_burnt,marker='o',color='b')
plt.plot(weight,marker='*',color='r')
plt.xlabel("calories_burnt")
plt.ylabel("weight")
plt.legend(labels=["calories_burnt","weight"],loc="lower right")
plt.xticks(ticks=[0,1,2,3,4],labels=["p1","p2","p3","p4","p5"]);
```

```python
plt.plot(calories_burnt,'ro--')
plt.plot(weight,'go-')
plt.xlabel("calories_burnt")
plt.ylabel("weight")
plt.legend(labels=["calories_burnt","weight"],loc="lower right")
plt.xticks(ticks=[0,1,2,3,4],labels=["p1","p2","p3","p4","p5"]);
```

```python
fig,ax =plt.subplots(nrows=1,ncols=2,figsize=(15,6))
ax[0].plot(calories_burnt,'go--')
ax[1].plot(weight)
ax[0].set_title('calories_burnt')
ax[1].set_title('weight')
ax[0].set_xticks(ticks=[0,1,2,3,4]);
ax[1].set_xticks(ticks=[0,1,2,3,4]);
ax[0].set_xticklabels(labels=['p1','p2','p3','p4','p5']);
ax[1].set_xticklabels(labels=['Q1','Q2','Q3','Q4','Q5']);
```

```python
fig,ax =plt.subplots(nrows=1,ncols=2,figsize=(15,6),sharey=True)
ax[0].plot(calories_burnt,'go--')
ax[1].plot(weight)
ax[0].set_title('calories_burnt')
ax[1].set_title('weight')
ax[0].set_xticks(ticks=[0,1,2,3,4]);
ax[1].set_xticks(ticks=[0,1,2,3,4]);
ax[0].set_xticklabels(labels=['p1','p2','p3','p4','p5']);
ax[1].set_xticklabels(labels=['Q1','Q2','Q3','Q4','Q5']);
```

```python
data_BM=pd.read_csv(r'C:\Users\Admin\Downloads\bigmart_data.csv')
data_BM= data_BM.dropna(how="any")
data_BM.head()
```

```python
price_by_item =data_BM.groupby('Item_Type').Item_MRP.mean()[:10]
price_by_item
```

```python
#Line Chart

price_by_item =data_BM.groupby('Item_Type').Item_MRP.mean()[:10]
price_by_item
x=price_by_item.index.tolist()
y=price_by_item.values.tolist()
plt.figure(figsize=(15,6))
plt.title("mean price for each item type")
plt.xlabel("Item type")
plt.ylabel("Mean Price")
plt.xticks(labels=x,ticks=np.arange(len(x)));
plt.plot(x,y)
```

```python
sales_by_Outlet_Size=data_BM.groupby('Outlet_Size').Item_Outlet_Sales.mean()
sales_by_Outlet_Size
```

```python
#Histogram

plt.title('Item MRP (price) distribution')
plt.xlabel("Item MRP")
plt.ylabel("Frequency")
plt.hist(data_BM['Item_MRP'],bins=20,color='skyblue');
```
